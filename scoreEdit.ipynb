```python
f = open("myNewRankedLists.txt","r")
f1 = f.readlines()
ans = {}
for x in f1:
    y = x.split()
    value = y[6]
    z = x.split('_')
    q = z[2]
    i = z[0]
    i = i.split()
    i = i[4]
    q = q.split()
    q = q[0]
    if q not in ans:
        ans[q] = []
    ans[q].append([y[6],i])
        
for key in ans.keys():
    print(key)
    cnt = 10
#     print(ans[key])
    ans[key].sort()
    #print(ans[key])
    val = []
    li = []
    for i in ans[key]:
        if i[0] not in val:
            cnt = cnt - 1
            val.append(i[0])
        li.append([i[1],cnt])
    print(li) 
```
